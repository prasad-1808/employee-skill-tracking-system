```python
import pandas as pd
import random
from faker import Faker

# Initialize Faker
fake = Faker()

# Predefined roles and their relevant skills
skills_mapping = {
    'junior data scientist': ['Python', 'SQL', 'Data Visualization'],
    'senior data scientist': ['Machine Learning', 'TensorFlow', 'Natural Language Processing'],
    'junior data engineer': ['Python', 'ETL', 'Data Warehousing'],
    'senior data engineer': ['Hadoop', 'Spark', 'Data Lakes'],
    'junior software engineer': ['JavaScript', 'HTML', 'CSS'],
    'senior software engineer': ['Microservices', 'DevOps', 'Cloud Computing'],
    'junior data analyst': ['Excel', 'Tableau', 'SQL'],
    'senior data analyst': ['Statistics', 'Predictive Analytics', 'Data Modeling']
}

# Courses mapped directly to their respective role outcomes
courses_mapping = {
    'junior data scientist': [
        'Introduction to Data Science', 
        'Python for Data Science', 
        'Data Analysis with Pandas'
    ],
    'senior data scientist': [
        'Advanced Machine Learning', 
        'Deep Learning Specialization', 
        'AI for Everyone'
    ],
    'junior data engineer': [
        'Introduction to Data Engineering', 
        'ETL Fundamentals', 
        'Building Data Pipelines'
    ],
    'senior data engineer': [
        'Big Data Technologies', 
        'Cloud Data Engineering', 
        'Data Architecture and Design'
    ],
    'junior software engineer': [
        'Web Development Bootcamp', 
        'JavaScript Basics', 
        'Introduction to React'
    ],
    'senior software engineer': [
        'Microservices Architecture', 
        'DevOps Essentials', 
        'Advanced JavaScript'
    ],
    'junior data analyst': [
        'Data Analysis with Excel', 
        'SQL Basics for Data Analysis', 
        'Tableau for Beginners'
    ],
    'senior data analyst': [
        'Advanced Data Visualization', 
        'Predictive Analytics in R', 
        'Business Intelligence Essentials'
    ]
}

# Function to generate course data
def generate_course_data(courses_mapping):
    course_data = []
    course_id_counter = 1  # Start course IDs from 1
    
    for role, courses in courses_mapping.items():
        for course in courses:
            course_record = {
                'course_id': course_id_counter,  # Use sequential ID
                'course_name': course,
                'course_duration': f"{random.randint(1, 12)} months",  # Random duration between 1 to 12 months
                'price': f"${random.randint(100, 2000)}",  # Random price between $100 and $2000
                'outcome_skill': random.choice(skills_mapping[role]),  # Random skill associated with the role
                'role_outcome': role  # Assign full role title based on course
            }
            course_data.append(course_record)
            course_id_counter += 1  # Increment the course ID counter

    return course_data

# Generate course data
course_data = generate_course_data(courses_mapping)

# Convert to DataFrame
df_courses = pd.DataFrame(course_data)

# Save to CSV
df_courses.to_csv('course_data.csv', index=False)

print('Course dataset generated and saved to course_data.csv')
```
