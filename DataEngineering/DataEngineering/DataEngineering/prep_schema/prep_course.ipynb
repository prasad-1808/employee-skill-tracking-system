```python
import pandas as pd
import numpy as np
```

```python
file_path = r"C:\Users\Admin\Desktop\JMD327_MainProject\JMD327_JElite\DataEngineering\DataEngineering\DataEngineering\raw_schema\raw_course.csv"
course = pd.read_csv(file_path)
course.head()
```

```python
course['price'] = course['price'].str.replace(r'\D', '', regex=True)

# Convert the cleaned price to an integer
course['price'] = course['price'].astype(int)
course.head()
```

```python
course['course_duration'] = course['course_duration'].str.replace(r'\D', '', regex=True)
course['course_duration'] = course['course_duration'].astype(int)
course.head()
```

```python

course.to_csv("prep_course.csv", index=False)
```
